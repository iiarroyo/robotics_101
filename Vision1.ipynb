```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2gray, gray2rgb
%matplotlib inline
#holahola
```

```python
path = os.path.join('images', 'lion.JPG')
lion = imread(path)
lion_grey = rgb2gray(lion)
```

# Lion

```python
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.imshow(lion), plt.axis('off')
plt.title('Colour lion', size=30)
plt.subplot(1, 2, 2)
plt.imshow(lion_grey, cmap=plt.cm.gray), plt.axis('off')
plt.title('Grey scale lion', size=30)
plt.show()
lion.shape
```

# Pepper

```python
path = os.path.join('images', 'peppers.png')
peppers = imread(path)
plt.figure(figsize=(20, 10))
plt.subplot(1, 4, 1)
plt.imshow(peppers), plt.axis('off'), plt.title('hola', size=30)
plt.subplot(1, 4, 2)
plt.imshow(peppers[..., 0], cmap=plt.cm.Reds), plt.axis('off'), plt.title('hola', size=30)
plt.subplot(1, 4, 3)
plt.imshow(peppers[..., 1], cmap=plt.cm.Greens), plt.axis('off'), plt.title('hola', size=30)
plt.subplot(1, 4, 4)
plt.imshow(peppers[..., 2], cmap=plt.cm.Blues), plt.axis('off'), plt.title('hola', size=30)
plt.show()
peppers.shape
```
